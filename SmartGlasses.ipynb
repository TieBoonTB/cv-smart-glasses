# Import modules

```python
from ultralytics import YOLO
import cv2
import os
```

# YOLO inferencing

```python
os.getcwd()
```

```python
video_path = "./input_video/20210125011541_202108.MP4"
```

```python
video_name = video_path.split("input_video/")[1]
video_name = video_name.split(".MP4")[0]
video_name
```

```python
# Load pre-trained YOLOv8 model
model = YOLO("yolov8n.pt")

# Open video
cap = cv2.VideoCapture(video_path)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Output video writer
output_folder = "./output_video"
output_path = f"{output_folder}/{video_name}_output.mp4"
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLOv8 on the frame
    # results = model(frame, save=True, project=output_folder)[0]
    results = model(frame)[0]

    # Draw results on the frame
    annotated_frame = results.plot()

    # Write annotated frame to output video
    out.write(annotated_frame)

    # Show frame
    # cv2.imshow('YOLOv8 Detection', annotated_frame)
    # if cv2.waitKey(1) & 0xFF == ord('q'):
    #     break

cap.release()
out.release()
cv2.destroyAllWindows()
```
